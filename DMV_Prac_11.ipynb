```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("City_Air_Quality.csv")
```

```python
df.head()
```

```python
# Display the first few rows and general info
print(df.head())
print(df.info())
```

```python
# Create a datetime column if 'date' and 'time' are separate
df['datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['datetime'], df['AQI'], color='purple', label='AQI')
plt.title("AQI Trend Over Time")
plt.xlabel("Date")
plt.ylabel("AQI Value")
plt.legend()
plt.show()
```

```python
# PM2.5 Trend
plt.figure(figsize=(12, 6))
plt.plot(df['datetime'], df['PM2.5'], color='red', label='PM2.5')
plt.title("PM2.5 Levels Over Time")
plt.xlabel("Date")
plt.ylabel("PM2.5 Level")
plt.legend()
plt.show()

# PM10 Trend
plt.figure(figsize=(12, 6))
plt.plot(df['datetime'], df['PM10'], color='blue', label='PM10')
plt.title("PM10 Levels Over Time")
plt.xlabel("Date")
plt.ylabel("PM10 Level")
plt.legend()
plt.show()

# CO Trend
plt.figure(figsize=(12, 6))
plt.plot(df['datetime'], df['CO'], color='green', label='CO')
plt.title("CO Levels Over Time")
plt.xlabel("Date")
plt.ylabel("CO Level")
plt.legend()
plt.show()
```

```python
# Set a daily frequency for comparison
daily_aqi = df.set_index('datetime').resample('D')['AQI'].mean().dropna()

plt.figure(figsize=(14, 7))
daily_aqi.plot(kind='bar', color='cyan')
plt.title("Daily Average AQI Values")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df[['AQI', 'PM2.5', 'PM10', 'CO']].boxplot()
plt.title("Distribution of AQI and Pollutant Levels")
plt.ylabel("Values")
plt.show()
```

```python
# Scatter Plot for AQI vs PM2.5
plt.figure(figsize=(8, 6))
plt.scatter(df['PM2.5'], df['AQI'], color='coral', alpha=0.5)
plt.title("AQI vs PM2.5 Levels")
plt.xlabel("PM2.5 Level")
plt.ylabel("AQI")
plt.show()

# Scatter Plot for AQI vs PM10
plt.figure(figsize=(8, 6))
plt.scatter(df['PM10'], df['AQI'], color='skyblue', alpha=0.5)
plt.title("AQI vs PM10 Levels")
plt.xlabel("PM10 Level")
plt.ylabel("AQI")
plt.show()

# Scatter Plot for AQI vs CO
plt.figure(figsize=(8, 6))
plt.scatter(df['CO'], df['AQI'], color='limegreen', alpha=0.5)
plt.title("AQI vs CO Levels")
plt.xlabel("CO Level")
plt.ylabel("AQI")
plt.show()
```
